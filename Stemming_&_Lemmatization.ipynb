```python
import nbformat

def get_notebook_content(notebook_path):
    with open(notebook_path, 'r', encoding='utf-8') as f:
        nb = nbformat.read(f, as_version=4)

    content = []
    for cell in nb.cells:
        if cell.cell_type == 'code':
            content.append(cell.source)
        elif cell.cell_type == 'markdown':
            content.append(f"'''{cell.source}'''")

    return '\n\n'.join(content)

notebook_path = # replace with your notebook path
print(get_notebook_content(notebook_path))
```

```python
import nltk

from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
words = ['run' ,'runner' , 'running', 'ran' , 'runs', 'easliy' , 'fairly']
```

```python
for word in words:
  print(word , "--> " , p_stemmer.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
s_stemmer = SnowballStemmer(language='english')
```

```python
words
```

```python
for word in words:
  print(word, "--->" , s_stemmer.stem(word))
```

```python
word ="consolingly"
print(f"porter stemmer :\n{word} ->" ,p_stemmer.stem(word))
print(f"\nsnowball stemmer :\n{word}" ,s_stemmer.stem(word))
```

```python
phrase = "I am meeting him tomorrow at the meeting"

for word in phrase.split():
  print(f"porter stemmer :{word} ->" ,p_stemmer.stem(word))

  print(f"snowball stemmer :{word}" ,s_stemmer.stem(word) , "\n\n")
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc1 = nlp(u'I am a runner in a race because I love to run since I ran today')

for token in doc1:
  print(token.text , "\t" , token.pos_ , "\t" , token.lemma , "\t" , token.lemma_)
```

```python
def show_lemmas(text):
  for token in text:
    print(f'{token.text:{12}} {token.pos_:{6}} {token.lemma:<{22}} {token.lemma_}')
```

```python
doc2 = nlp(u'I saw eighteen mice today!')
show_lemmas(doc2)
```

```python
doc3 = nlp(u"I am meeting him tommorrow at the meeting.")

show_lemmas(doc3)
```

```python
doc = nlp(u"That's and enormous automobile")

show_lemmas(doc)
```
